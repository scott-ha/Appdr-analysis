```python
# df_dev_platform_euclidean_work4 = df_dev_platform_euclidean_num.loc[:, ['Android', 'Admin page', 'Mobile Web']]

df_cluster = df_dev_platform_euclidean_num.loc[:, ['Android', 'Admin page']]
df_cluster
```

```python
# developer clustering
# using df_dev_platform_euclidean_num 

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

model = KMeans(n_clusters=2, algorithm='auto')
model.fit(df_cluster)
predict = pd.DataFrame(model.predict(df_cluster))
predict.columns=['predict']
```

```python
d = pd.concat([df_cluster, predict], axis=1)
d
```

```python
# platforms = ['Android','IOS','Admin page','Mobile Web','RWD Web','Server','PC Web','DB']

plt.scatter(d['Android'], d['Admin page'], c=d['predict'],alpha=0.5)

centers = pd.DataFrame(model.cluster_centers_,columns=['Android','IOS'])

center_x = centers['Android']
center_y = centers['IOS']

plt.scatter(center_x,center_y,s=50,marker='D',c='r')

plt.show()

```
