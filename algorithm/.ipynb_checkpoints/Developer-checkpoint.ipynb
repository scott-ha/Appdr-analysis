```python
appdr_data = pd.read_csv('../../data/20200323_issues(utf-8).csv', encoding='utf-8')
appdr_data
```

```python
# preprocessing

# column 공백 제거
df = appdr_data.rename(columns={'#':'issue_num',
                        'Parent task':'Parent_task', 
                        'Target version':'Target_version', 
                        'Start date':'Start_date', 
                        'Due date':'Due_date', 
                        'Estimated time':'Estimated_time', 
                        'Total estimated time':'Total_estimated_time', 
                        'Spent time':'Spent_time', 
                        'Total spent time':'Total_spent_time',
                        '% Done':'%_Done',
                        'Last Updated by':'Last_Updated_by', 
                        'Related issues':'Related_issues', 
                        'HQ time':'HQ_time', 
                        'Resolution Incentive':'Resolution_Incentive', 
                        'Act Date':'Act_Date', 
                        'Message to customer':'Message_to_customer', 
                        'Subject (for customer)':'Subject(for_customer)', 
                        'Business Analystic(BA)':'Business_Analystic(BA)', 
                        'Source management':'Source_management', 
                        'Fix bugs time':'Fix_bugs_time',
                        'Front-end':'Front_end',
                        'Server-side':'Server_side'})
```

```python
# preprocessing 
# 앞뒤 공백 제거
# Assignee Duc Tran AOS -> Duc_Tran_AOS

df['Assignee'] = df['Assignee'].str.strip()
df['Assignee'] = df['Assignee'].str.replace(" ", "_")
```

```python
df
```

```python
# preprocessing
# null 
# Assignee, Skill, Platform, Estimated_time
# Assignee contains Appdr & AppDr
Assign_Appdr = pd.Series(df.Assignee)

df = df[pd.isnull(df.Assignee) != True]
df = df[pd.isnull(df.Skill) != True]
df = df[pd.isnull(df.Platform) != True]
df = df[Assign_Appdr.str.contains('Appdr') != True]
df = df[Assign_Appdr.str.contains('AppDr') != True] 
df = df[pd.isnull(df.Estimated_time) != True]
                
df.shape
```

```python
df
```

```python
df.Assignee.unique()
```

```python
developers = df.Assignee.unique()
```

```python
df.Platform.value_counts()
```

```python
platforms = ['Android','IOS','Admin page','Mobile Web','RWD Web','Server','PC Web','DB']
```

```python
#############################################
# 특정 개발자가 Platform의 어떤 스택을 보유하고 있는지 #
#############################################

# dev_df
# 개발자별 분류 -> 어떤 플랫폼?
# developers

# DataFrame for each developers
# df_Appdr_JJ = dev_df.query('Assignee=="Appdr JJ"')

# df_ex3 = pd.DataFrame(columns=(platforms), data=[list_ex3], index=[developers])

list_platform_count=[]
df_dev_platform = pd.DataFrame(columns=(platforms))

for i in developers:
  df_temp = df.query('Assignee==@i')
#   print(df_temp)
  for j in platforms:
    list_platform_count.append(df_temp.Platform.str.count(j).sum())
  print(list_platform_count)
  df_temp2 = pd.DataFrame(columns=(platforms), data=[list_platform_count], index=[i])
  df_dev_platform = df_dev_platform.append(df_temp2)
  list_platform_count=[]
  
df_dev_platform
```

```python
df_dev_platform.shape
```

```python
df.Skill.value_counts()
```

```python
# Skill
# Android Studio, Swift, HTML/CSS, PHP, PHP - Larabel, Javascript
# React Native, NodeJS, Kotlin, React, MySQL, AngularJS, Ruby on Rail, VueJS
# Python Django, Server AWS, PHP - CodeIgniter, POSTGre, Server Cafe24, Python Flask
# Ubuntu, Linux, ASP, .NET, MSSQL, object-c, JSP/Java, Docker, Wordpress, CentOS,
# Appcenter (for app auto Distribution), Java Spring, NodeJS Express, PHP - GNU board
```

```python
skills = ['Android Studio', 'Swift', 'HTML/CSS', 'PHP', 'PHP - Larabel', 
          'Javascript', 'React Native', 'NodeJS', 'Kotlin', 'React', 'MySQL', 'AngularJS', 
          'Ruby on Rail', 'VueJS', 'Python Django', 'Server AWS', 'PHP - CodeIgniter', 
          'POSTGre', 'Server Cafe24', 'Python Flask', 'Ubuntu', 'Linux', 'ASP', '.NET', 'MSSQL', 
          'object-c', 'JSP/Java', 'Docker', 'Wordpress', 'CentOS', 'Appcenter (for app auto Distribution)', 
          'Java Spring', 'NodeJS Express', 'PHP - GNU board']
```

```python
skills
```

```python
#############################################
# 특정 개발자가 Skill의 어떤 스택을 보유하고 있는지 #
#############################################

# dev_df
# 개발자별 분류 -> 어떤 플랫폼?
# developers

# DataFrame for each developers
# df_Appdr_JJ = dev_df.query('Assignee=="Appdr JJ"')

# df_ex3 = pd.DataFrame(columns=(platforms), data=[list_ex3], index=[developers])

list_skill_count=[]
df_dev_skill = pd.DataFrame(columns=(skills))

for i in developers:
  df_temp = df.query('Assignee==@i')
#   print(df_temp)
  for j in skills:
    list_skill_count.append(df_temp.Skill.str.count(j).sum())
  print(list_skill_count)
  df_temp2 = pd.DataFrame(columns=(skills), data=[list_skill_count], index=[i])
  df_dev_skill = df_dev_skill.append(df_temp2)
  list_skill_count=[]
  
df_dev_skill
```

```python
df_dev_skill.shape
```

```python
# df_dev_platform , df_dev_skill
# (37,8) (37,34)
df_developer = pd.merge(df_dev_platform, df_dev_skill, how="outer", left_index=True, right_index=True)
```
