```python
# 전처리한 데이터로 치환 필요함
issue_data = pd.read_csv('../../data/pre_data_200513.csv', encoding='utf-8')
issue_data
```

```python
list_a = []
column = ['issue', 'issue_count']

df_p_issue = pd.DataFrame(columns=(column))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    c = df_temp.issue_num.values
    d = len(df_temp.issue_num.values)
    data = [c,d]
    df_temp2 = pd.DataFrame(columns=(column), data=[data], index=[i])
    df_p_issue = df_p_issue.append(df_temp2)
    
df_p_issue
```

```python
list_Etime_count=[]
E_time_columns = ['ET(0-2)','ET(2-4)','ET(4-6)','ET(6-20)','ET(20-)', 'ET_Sum(issue_count)', 'ET_Sum']
df_est = pd.DataFrame(columns=(E_time_columns))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    for j in E_time_columns:
        E_range0 = len(df_temp['Estimated_time'][df_temp['Estimated_time']<=2])
        E_range1 = len(df_temp['Estimated_time'][(df_temp['Estimated_time']>2) & (df_temp['Estimated_time']<=4)])
        E_range2 = len(df_temp['Estimated_time'][(df_temp['Estimated_time']>4) & (df_temp['Estimated_time']<=6)])
        E_range3 = len(df_temp['Estimated_time'][(df_temp['Estimated_time']>6) & (df_temp['Estimated_time']<=20)])
        E_range4 = len(df_temp['Estimated_time'][df_temp['Estimated_time']>20])
        ET_Sum = sum(df_temp['Estimated_time'])
        E_sum = E_range0 + E_range1 + E_range2 + E_range3 + E_range4 
        list_Etime_count=[E_range0, E_range1, E_range2, E_range3, E_range4, E_sum, ET_Sum]
    print(list_Etime_count)
    df_temp2 = pd.DataFrame(columns=E_time_columns, data=[list_Etime_count], index=[i])
    df_est = df_est.append(df_temp2)
    list_Etime_count=[]
```

```python
df_est
```

```python
df_p_issue = pd.merge(df_p_issue, df_est, how="outer", left_index=True, right_index=True)
```

```python
df_p_issue
```

```python
# Project에 할당한 개발자?

list_p_developer = []
d_columns = ['developer', 'num_developers']
df_developer = pd.DataFrame(columns=(d_columns))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    c = df_temp.Assignee.values
    developers = pd.unique(c)
    # 개발자 수 
    num_developers = len(developers)
    data = [developers, num_developers]
    df_temp2 = pd.DataFrame(columns=(d_columns), data=[data], index=[i])
    df_developer = df_developer.append(df_temp2)

df_developer
```

```python
df_p_issue = pd.merge(df_p_issue, df_developer, how="outer", left_index=True, right_index=True)
```

```python
platforms = ['Android','IOS','Admin page','Mobile Web','RWD Web','Server','PC Web','DB']
```

```python
# Project에 사용된 platform

list_platform = []
d_columns = ['used_platform', 'num_platform']
df_p = pd.DataFrame(columns=(d_columns))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    c = df_temp.Platform.values
    used_platform = pd.unique(c)
    num_platform = len(used_platform)
    data = [used_platform, num_platform]
    df_temp2 = pd.DataFrame(columns=(d_columns), data=[data], index=[i])
    df_p = df_p.append(df_temp2)

df_p
```

```python
# Project에 사용된 Platform

list_platform_count=[]
df_platform = pd.DataFrame(columns=(platforms))
    
for i in p_name:
    df_temp = issue_data.query('Project==@i')
    for j in platforms:
        list_platform_count.append(df_temp.Platform.str.count(j).sum())
    df_temp2 = pd.DataFrame(columns=(platforms), data=[list_platform_count], index=[i])
    df_platform = df_platform.append(df_temp2)
    list_platform_count=[]
    
df_platform = pd.merge(df_p, df_platform, how="outer", left_index=True, right_index=True)
df_platform
```

```python
df_p_issue = pd.merge(df_p_issue, df_platform, how="outer", left_index=True, right_index=True)
df_p_issue
```

```python
skills = ['Android Studio', 'Swift', 'HTML/CSS', 'PHP', 'PHP - Larabel', 
          'Javascript', 'React Native', 'NodeJS', 'Kotlin', 'React', 'MySQL', 'AngularJS', 
          'Ruby on Rail', 'VueJS', 'Python Django', 'Server AWS', 'PHP - CodeIgniter', 
          'POSTGre', 'Server Cafe24', 'Python Flask', 'Ubuntu', 'Linux', 'ASP', '.NET', 'MSSQL', 
          'object-c', 'JSP/Java', 'Docker', 'Wordpress', 'CentOS', 'Appcenter (for app auto Distribution)', 
          'Java Spring', 'NodeJS Express', 'PHP - GNU board']
```

```python
# Project에 사용된 skill

list_skill = []
d_columns = ['used_skill', 'num_skill']
df_s = pd.DataFrame(columns=(d_columns))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    c = df_temp.Skill.values
    used_skill = pd.unique(c)
    num_skill = len(used_skill)
    data = [used_skill, num_skill]
    df_temp2 = pd.DataFrame(columns=(d_columns), data=[data], index=[i])
    df_s = df_s.append(df_temp2)

df_s
```

```python
# Project에 사용된 Skill

list_skill_count=[]
df_skill = pd.DataFrame(columns=(skills))

for i in p_name:
    df_temp = issue_data.query('Project==@i')
    for j in skills:
        list_skill_count.append(df_temp.Skill.str.count(j).sum())
#     print(list_skill_count)
    df_temp2 = pd.DataFrame(columns=(skills), data=[list_skill_count], index=[i])
    df_skill = df_skill.append(df_temp2)
    list_skill_count=[]
    
df_skill = pd.merge(df_s, df_skill, how="outer", left_index=True, right_index=True)
df_skill
```

```python
df_p_issue = pd.merge(df_p_issue, df_skill, how="outer", left_index=True, right_index=True)
df_p_issue
```
