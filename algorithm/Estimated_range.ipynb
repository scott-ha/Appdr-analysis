```python
# list * list EX

list1 = [5,4,3,2,1] 
list2 = [5,4,3,2,1]
result = []

for num1, num2 in zip(list1, list2):
    result.append(num1 * num2)
print(result)

# plus 
in_list1 = 0
in_list2 = 1

for i in result:
    print(i)
    in_list1 += i
    in_list2 *= i
    
print(in_list1, in_list2)
```

```python
# Weight METHOD 1
# -1
# CASE : ET(0-2)

et_weight = [5,4,3,2,1]
et_result = []
et_result_sum = 0
wr_1 = ['weight(0-2)']
df_est_range1 = pd.DataFrame(columns=(wr_1))

for i in developers:
    df_tmp = df_est.query('developer==@i')
    list_tmp = df_tmp.values.tolist()
    list_tmp = list_tmp[0]
    for num1, num2 in zip(et_weight, list_tmp):
        et_result.append(num1 * num2)
    for k in et_result:
        et_result_sum += k
    print(et_result, et_result_sum)
    df_tmp2 = pd.DataFrame(columns=wr_1, data=et_result_sum, index=[i])
    df_est_range1 = df_est_range1.append(df_tmp2)
    et_result=[]
    et_result_sum = 0
    
df_est_range1.index.names = ['developer']
df_est_range1
```
