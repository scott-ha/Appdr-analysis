```python
# df : proprecessing data null 
# Assignee null
df = appdr_data[pd.isnull(appdr_data.Assignee) != True]
```

```python
df.shape
```

```python
# Skill null
df = df[pd.isnull(df.Skill) != True]
```

```python
df.shape
```

```python
# Assignee contains Appdr ~
Assign_Appdr = pd.Series(df.Assignee)
df = df[Assign_Appdr.str.contains('Appdr') != True]
```
