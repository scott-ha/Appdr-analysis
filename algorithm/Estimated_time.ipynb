```python
# 절대값을 가지는 경우의 계산
df = Developer_data[['Developer', 'Estimated_time_sum', 'Estimated_time_avg', 'issue_count']]
# Developers
developers = df.Developer.unique()
df
```

```python
# project_estimated_time = 10
# avg_est_time = df['Estimated_time_avg'].tolist()
# avg_est_time
```

```python
project_estimated_time = 57
avg_est_time = df['Estimated_time_avg'].tolist()
distance_list = []

for i in avg_est_time:
    distance = project_estimated_time -i
    distance = abs(distance)
    distance_list.append(distance)
    
# distance_list
est_distance = ['est_distance']
df_est_distance = pd.DataFrame(columns=(est_distance), data=distance_list, index=developers)

df_est_distance
```
